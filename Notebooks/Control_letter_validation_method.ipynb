```python
#LDQ-LOQ

import matplotlib.pyplot as plt
import pandas as pd

# Datos de respuesta proporcionados
datos = {
    '1 ppm': [1.014910714, 0.982857143, 0.974375, 0.975982143, 0.965089286, 0.915625, 
              0.921160714, 0.920178571, 0.916339286, 0.872232143],
    '0.5 ppm': [0.500089286, 0.489821429, 0.480982143, 0.470267857, 0.466875, 0.46125, 
                0.465089286, 0.464375, 0.475803571, 0.459464286],
    '0.25 ppm': [0.271607143, 0.285357143, 0.283392857, 0.274910714, 0.274642857, 0.282232143, 
                 0.273125, 0.275625, 0.275089286, 0.272321429],
    '0.2 ppm': [0.16, 0.142, 0.137, 0.135, 0.13, 0.127, 0.13, 0.129, 0.129, 0.132],
    '0.1 ppm': [0.09, 0.07, 0.069, 0.064, 0.059, 0.066, 0.062, 0.061, 0.064, 0.063],
    '0.05 ppm': [0.053, 0.03, 0.026, 0.02, 0.018, 0.019, 0.018, 0.019, 0.017, 0.017]
}

# Convertimos a DataFrame y calculamos estadísticas
df_concentraciones = pd.DataFrame(datos)
df_stats = df_concentraciones.agg(['mean', 'std']).transpose()
df_stats['concentration'] = [float(conc.replace(' ppm', '')) for conc in df_stats.index]

# Corregimos aquí el nombre de la columna a 'std' que es el nombre correcto
df_stats['LSC'] = df_stats['mean'] + 3 * df_stats['std']  # Límite Superior de Control (Upper Control Limit, UCL)
df_stats['LIC'] = df_stats['mean'] - 3 * df_stats['std']  # Límite Inferior de Control (Lower Control Limit, LCL)

# Inicializamos la figura de Matplotlib
plt.figure(figsize=(10, 6))

# Graficamos el promedio de las respuestas para cada concentración
plt.plot(df_stats['concentration'], df_stats['mean'], 'o-', label='Average response')

# Graficamos los límites de control superior e inferior para cada punto de concentración
for conc in df_stats.index:
    plt.plot(df_stats.loc[conc, 'concentration'], df_stats.loc[conc, 'LSC'], 'r--')
    plt.plot(df_stats.loc[conc, 'concentration'], df_stats.loc[conc, 'LIC'], 'b--')

# Añadimos líneas horizontales para los límites de control general
plt.axhline(y=df_stats['LSC'].max(), color='red', linestyle='--', label='Upper control limit (UCL)')
plt.axhline(y=df_stats['LIC'].min(), color='blue', linestyle='--', label='Lower control limit (LCL)')

# Añadimos línea horizontal para el LOQ del método basado en el % de recuperación
loq_metodo = '0.25 ppm'  # Esto se determinó previamente en la conversación
loq_metodo_value = float(loq_metodo.replace(' ppm', ''))
plt.axhline(y=loq_metodo_value, color='green', linestyle='-', label=f'LOQ of the method ({loq_metodo_value} ppm)')

# Añadimos etiquetas y título al gráfico
plt.xlabel('Concentration (ppm)')
plt.ylabel('Method response (ppm)')
plt.title('Control Chart for Method Validation-LOD-LOQ')
plt.legend()

# Añadimos la cuadrícula al gráfico
plt.grid(True)

# Mostramos el gráfico
plt.savefig("../Result/Methods_validation/Figuras/PNG/figura1.png", dpi=300, bbox_inches='tight')  # Movido antes de show()
plt.savefig("../Result/Methods_validation/Figuras/PDF/figura1.pdf", dpi=300, bbox_inches='tight')  # Movido antes de show()
plt.show()
```

```python
#Precisión
import pandas as pd
from scipy.stats import variation

# Cargando el archivo Excel
file_path = 'C:/Users/User/Documents/Documentos_Tesis/Estadistica/Validacion de metodo/Presicion.xlsx'
data = pd.read_excel(file_path)

# Convirtiendo la columna 'Dia' al formato de fecha
data['Dia'] = pd.to_datetime(data['Day']).dt.date

# Calculando el coeficiente de variación porcentual (% CV) para la repetibilidad y precisión intermedia

# Repetibilidad (CV diario por rango)
cv_repetibilidad = data.groupby(['Day', 'Amount']).agg({'Concentration': lambda x: variation(x, axis=0) * 100})

# Precisión intermedia (CV total por rango)
cv_intermedia = data.groupby('Amount').agg({'Concentration': lambda x: variation(x, axis=0) * 100})

# Mostrando los resultados
print("Coefficient of Variation for Repeatability:")
print(cv_repetibilidad)
print("\nCoefficient of Variation for Intermediate Precision:")
print(cv_intermedia)
```

```python
#Carta control precision

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import variation

# Carga y preparación de los datos
file_path = 'C:/Users/User/Documents/Documentos_Tesis/Estadistica/Validacion de metodo/Presicion.xlsx'
data = pd.read_excel(file_path)
data['Day'] = pd.to_datetime(data['Day']).dt.date

# Calculando medias y rangos
medias = data.groupby(['Day', 'Amount'])['Concentration'].mean().unstack()
rangos = data.groupby(['Day', 'Amount'])['Concentration'].agg(np.ptp).unstack()

# Función modificada
def crear_carta_control_con_etiquetas(medias, rangos, titulo):
    dias = medias.index
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10), sharex=True)

    # Carta de Medias (X-bar)
    ax1.plot(dias, medias, marker='o', label='Daily average')
    ax1.set_title(f'Mean control chart for {titulo}')
    ax1.set_ylabel('Concentration Average')

    # Líneas de control para la carta de Medias
    media_total = medias.mean()
    rango_promedio = rangos.mean()
    lcl_x = media_total - rango_promedio * 1.96 / np.sqrt(4)  # Límite de control inferior
    ucl_x = media_total + rango_promedio * 1.96 / np.sqrt(4)  # Límite de control superior
    ax1.axhline(media_total, color='green', linestyle='--', label='Media total')
    ax1.axhline(lcl_x, color='red', linestyle='--', label='Lower control limit')
    ax1.axhline(ucl_x, color='red', linestyle='--', label='Upper control limit')
    ax1.legend()

    # Carta de Rangos (R)
    ax2.plot(dias, rangos, marker='o', color='orange', label='Daily range')
    ax2.set_title(f'Rank Control Chart for {titulo}')
    ax2.set_ylabel('% Concentration Range')
    ax2.set_xlabel('Day')

    # Líneas de control para la carta de Rangos
    lcl_r = 0  # Límite de control inferior siempre es 0 para rangos
    ucl_r = rango_promedio * 2.114  # Límite de control superior
    ax2.axhline(rango_promedio, color='green', linestyle='--', label='Average range')
    ax2.axhline(lcl_r, color='red', linestyle='--', label='Lower control limit')
    ax2.axhline(ucl_r, color='red', linestyle='--', label='Upper control limit')
    ax2.legend()

    plt.tight_layout()
    # Guardar la figura
    plt.savefig(
        f"../Result/Methods_validation/Figuras/PDF/carta_control_{titulo}.pdf", 
        dpi=300, 
        bbox_inches='tight'
    )
    plt.show()
# Creando cartas de control con etiquetas para cada rango
for cantidad in medias.columns:
    crear_carta_control_con_etiquetas(medias[cantidad], rangos[cantidad], cantidad)
```
